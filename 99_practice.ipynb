```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
df.head()
df.tail()
# df['species'].value_counts()
```

```python
# 5000g 이상의 펭귄 출력
con1 = df['body_mass_g'] >= 5000
df[con1]
```

```python
# Gentoo종의 flipper_length 평균
con1 = df['species'] == 'Gentoo'

Gentoo_mean = df.loc[con1, 'flipper_length_mm'].mean()
Gentoo_mean
```

```python
# Female Chinstrap종의 bill_length 높은순 정렬
#df['species'].value_counts() # 범주 분포 확인

con1 = df['species'] == 'Chinstrap'
con2 = df['sex'] == 'Female'

df.loc[con1 & con2, 'bill_length_mm'].sort_values().head(10)
```

```python
# species(종)별 성별 구분하여 평균 몸무게 출력
# => ex) Adelie, Female / Adelie, male / Gentoo, Female...

# Adelie
con1 = df['species'] == 'Adelie'
con1_1 = df.loc[con1, 'sex'] == 'Male'
con1_2 = df.loc[con1, 'sex'] == 'Female'

Adeline_M_mean = df.loc[con1 & con1_1, 'body_mass_g'].mean()
Adeline_F_mean = df.loc[con1 & con1_2, 'body_mass_g'].mean()

# Gentoo
con2 = df['species'] == 'Gentoo'
con2_1 = df.loc[con2, 'sex'] == 'Male'
con2_2 = df.loc[con2, 'sex'] == 'Female'

Gentoo_M_mean = df.loc[con2 & con2_1, 'body_mass_g'].mean()
Gentoo_F_mean = df.loc[con2 & con2_2, 'body_mass_g'].mean()

# Chinstrap
con3 = df['species'] == 'Chinstrap'
con3_1 = df.loc[con3, 'sex'] == 'Male'
con3_2 = df.loc[con3, 'sex'] == 'Female'

Chinstrap_M_mean = df.loc[con3 & con3_1, 'body_mass_g'].mean()
Chinstrap_F_mean = df.loc[con3 & con3_2, 'body_mass_g'].mean()

print(f'아델리 수컷: {Adeline_M_mean}g')
print(f'아델리 암컷: {Adeline_F_mean}g')
print('---------')
print(f'젠투 수컷: {Gentoo_M_mean}g')
print(f'젠투 암컷: {Gentoo_F_mean}g')
print('---------')
print(f'턱끈 수컷: {Chinstrap_M_mean}g')
print(f'턱끈 암컷: {Chinstrap_F_mean}g')
```

```python
# bill_depth 결측치 채우기
# => 같은종 같은섬의 평균으로
con = df['bill_depth_mm'].isnull()
df[con]
```

```python
con1 = df['species'] == 'Adelie'
Adeline_mean = df.loc[con1, 'bill_depth_mm'].mean()

con2 = df['species'] == 'Gentoo'
Gentoo_mean = df.loc[con2, 'bill_depth_mm'].mean()

df.loc[con1, 'bill_depth_mm'] = df.loc[con1, 'bill_depth_mm'].fillna(Adeline_mean)
df.loc[con2, 'bill_depth_mm'] = df.loc[con2, 'bill_depth_mm'].fillna(Gentoo_mean)
```
