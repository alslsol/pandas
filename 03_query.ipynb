## indexing, slicing, 조건 필터링

```python
# 인덱싱: 인덱스 접근할 때는 Series 형태
df.loc[3] # df의 3번째 위치에 있는 데이터 출력해달란 것 >> 내가 가지고 오고 싶은 데이터 위치 찾을 때 사용
    # loc 통해 인덱싱 가능
# [위치 인덱싱, 컬럼네임] 
df.loc[3, 'class'] # df의 3번째 사람의 class 정보 출력
df.loc[3, ['class', 'age']] # 3번째 사람의 class와 age 정보 출력

df['class'] # 대괄호 하나 >> 시리즈로 반환해줌
df[['class']] # 대괄호 2개 >> 데이터 프레임으로 반환
```

```python
# 슬라이싱: 슬라이싱 접근할 때는 data frame 형태
df.loc[2:5] # 2~5번째 사람까지 정보 출력 >> 파이썬과 달리 마지막 데이터도 포함함
df.loc[2:5, ['pclass', 'who']] # 2-5번째의 pclass, who 정보만 출력
```

```python
df.loc[:3, 'class':'deck']
```

```python
condition = df['who'] == 'man' # 비교연산자: T/F로 뭉쳐진 데이터셋 만들어줌 >> 이를 condition에 저장해서 활용할 것
df[condition].head() # who에 man만 필터링됨
```

```python
df.loc[condition].head()
```

```python
df.loc[condition]['age'] # 남자만 뽑은 정보 중, 나이 정보만 뽑기 >> 필터링한 정보 내에서 loc 통해 인덱스 접근 가능
# 컬럼 접근할 때는, 
df.loc[condition, 'age'] = 100 # 남자 고른 데이터에서, age를 전부 100으로 채워달라는 것
df.head() # 조건 필터링 한 뒤, 일괄적으로 데이터 수정할 때 활용
```

```python
# 요금 30불 이상 낸 사람 필터링

cond1 = df['fare'] > 30

# 여자만 고르기
cond2 = df['who'] == 'woman'
```

```python
# 2개 정보 연결하기

df.loc[cond1 & cond2] # 왼/오 조건 모두 만족하는 경우만 필터링 >> 30불 이상 지불한 여자만 고르기
```

```python
# 2개 조건 중 하나라도 맞는 경우: 여자이거나 or 30불 낸 사람이거나, 둘다거나
df.loc[cond1 | cond2].head()
```

```python
df.iloc[1, 3] # iloc: 컬럼 이름이 아닌 숫자로 접근할 때 사용 >> 1번째 사람의 3번째 데이터 출력
# iloc: 인덱스 접근이므로 초과/미만O
```

```python
df.iloc[1:3, 2:5]
```

```python
# where()
cond = df['fare'] < 30 # 요금 30불 미만인 사람 필터링
df['fare'].where(cond, 1) # 30불보다 작으면 true 나오는데, 그걸 1로 바꾼단 것
# 일괄 변경할 때 사용
```

```python
#연습
# 내가 지불한 최종 금액과 팁 얼마 줬는지 대한 정보 >> 
tips = sns.load_dataset('tips') # seaborn이 기본적으로 가진 csv 파일 불러온 것
tips
```

```python
# 남성 흡연자 중 팁을 가장 많이 낸 순서로 내림차순 정렬

cond1 = tips['sex'] == 'Male'
cond2 = tips['smoker'] == 'Yes'

tips.loc[cond1 & cond2].sort_values('tip', ascending=False).head(10)

df1 = tips.loc[cond1 & cond2]
df1.sort_values('tip', ascending=False).head(10)
```

```python
# 식사 인원이 5명 이상인 저녁 테이블
cond1 = tips['size'] >= 5
cond2 = tips['time'] == 'Dinner'

# 동일코드: cond1 = tips['size'] >= 5 & tips['time'] == 'Dinner'

# 흡연자가 없는 주말 테이블
cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sun', 'Sat']) # day 컬럼 안에 Sun, Sat 둘 중 하나 있는 경우만 필터링

total = tips.loc[(cond1 & cond2) & (cond3 & cond4)], ['total_bil', 'tip']
# total_bill, tip 출력하기
total
# tips.loc[(cond1 & cond2) & (cond3 & cond4)][['total_bill', 'tip']]
# tips.loc[(cond1 & cond2) & (cond3 & cond4), ['total_bill', 'tip']]
```
