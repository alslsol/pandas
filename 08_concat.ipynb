```python
import pandas as pd

df1 = pd.read_csv('data/sales_data1.csv')
df2 = pd.read_csv('data/sales_data2.csv')

df1.head(2) # 상반기 정보
```

```python
df2.head(2) # 하반기 정보
```

```python
# 분리된 2개 데이터 프레임 하나로 합치기
# 컬럼 갯수는 같지만, 행 갯수는 다름 >> 2개 데이터를 행 방향으로(위아래)로 이어붙일 수 있음

df1.shape
```

```python
df2.shape
```

```python
# 데이터 프레임 합치기
df = pd.concat([df1, df2], ignore_index=True)
# 근데 인덱스가 240까지가X, df1 인덱스 번호 + df2 인덱스 번호 >> 이렇게 연결돼서 인덱스 중복됨
    ## 그래서 옵션 지정해서, 인덱스 번호 새로 만들기 해야 함: ignore_index=True
    # reset_index와 동일한 기능함
df.tail()
```

```python
# iloc로 해야 인덱스로 위치 접근 가능
# 컬럼 5개씩 좌우 분리해보기
sales1 = df1.iloc[:, :5] # : 전체 행 선택, :5 :첫번째부터 5번째 컬럼까지 선택하는 것
sales2 = df2.iloc[:, :5]
```

```python
# 좌우로 분리된 2개 df 하나로 합치기
pd.concat([sales1, sales2], axis=1) # axis=1: 왼쪽 세로 열에 sales2 붙여서 좌우로 합치겠다는 것
```

## merge()

```python
info = {
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}

money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
# dict을 df로 만들기
df1 = pd.DataFrame(info) # 인적 사항
df2 = pd.DataFrame(money) # 연봉 정보
# 마이클조던, 타이거우즈는 공통X, 각 df에 따로 존재함

# 2개 데이터 분리돼 있는데 >> 2개 연결하기
```

```python
pd.merge(df1, df2, how='inner') # 교집합만 남음 >> R에서 inner_join의 역할인 셈, 합집합:outer_join, 차집합: left/right_join
    # how=left, right,outer,inner >> 해서 어떻게 집합 만들건지 옵션 줄 수 있음
```

```python
pd.merge(df1, df2, how='outer') # 합집합 만듦 >> 대신 없는 데이터는 자동으로 결측으로 설정됨
```

```python
pd.merge(df1, df2, how='right')
```

```python
info = {
    '이름': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}

money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']}

df1 = pd.DataFrame(info)
df2 = pd.DataFrame(money)
```

```python
pd.merge(df1, df2, left_on='이름', right_on='고객명') # df 이름-고객명 서로 다르기 때문에, 직접 지정해서 2개를 기준으로 하나로 병합해달란 것 >> 사전에 컬럼 이름 동일하게 변경하는 게 더 쉬움
```

## 실습

```python
movies = pd.read_csv('data/movies.csv')
reviews = pd.read_csv('data/reviews.csv')
# movies.shape
# reviews.shape
# movies, reviews 서로 행/열 다름 >> 이를 교집합/합집합 등으로 합치고 싶을 때 merge 쓰는 것

movies.head(1) # movieId라는 첫 key 값 동일한 것만 교집합된 것
```

```python
# 2018년에 개봉한 영화의 모든 리뷰를 보고 싶다면: 2개 merge해서 접근해야 함

movie_review = pd.merge(movies, reviews) # 기본값: inner
movie_review.head(2) # 각 영화에 대한 리뷰 바로 아래 붙음: 영화 정보 + 영화 리뷰 >> 
```

```python
cond = movie_review['movieYear'] == 2005 # 2005년 개봉 영화만 필터링
movie_review.loc[cond, 'rating'].mean() # 2005년 개봉 영화 중 영화 평점만 출력해 평균 매긴 것
```
