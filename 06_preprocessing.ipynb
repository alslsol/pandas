```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
df_copy = df.copy()
```

```python
df_copy['VIP'] = False # VIP라는 컬럼 추가, 값은 False 넣어 생성
df_copy.head()
```

```python
# 글자끼리 연산(이어붙이기)해서 컬럼 생성
df_copy['gender'] = df_copy['sex'] + '-' + df_copy['who']
df_copy.head()
```

```python
df_copy.drop('VIP', axis=1).head()
# axis=1: 세로줄 의미
```

## 자전거 데이터 활용

```python
df = pd.read_csv('data/seoul_bicycle.csv')
df.head()
df.info()
# 대여일자: object 타입임 >> datetime이 아닌, 문자열로 기입돼 있는 것 >> 판다스가 인식 가능한 datetime 형태로 수정 필요
```

```python
df_copy = df.copy()
```

```python
# to_datetime(): datetime 객체로 변경해줌
df_copy['대여일자'] = pd.to_datetime(df_copy['대여일자'], format='%b-%d-%Y')
    # 지금 format=%b-%d-%Y 이런 형식으로 데이터 들어 있는데, 이를 파이썬이 인식하는 datatime 형태로 바꾸기
df_copy.info()
```

```python
# 특정 날짜만 필터링
cond = df_copy['대여일자'].dt.dayofweek == 3
df_copy.loc[cond]
```

```python
df_copy.head()
```

```python
# 운동량 컬럼의 값을 숫자로 바꾸기
pd.to_numeric(df_copy['운동량'])
```

```python
df_copy.loc[2344] # 해당 행에 \N 이라는 문자열 들어 있어서 숫자 전환 안 되는 것
```

```python
# 에러날 때 어떻게 할 지를 인자 설정하기
df_copy['운동량'] = pd.to_numeric(df_copy['운동량'], errors='coerce') # 숫자로 못 바꾸는 값은 nan으로 바꾸도록 설정
df_copy['운동량'].info # 숫자형으로 변경 완료
```

### cut(컬럼명, 리스트명) 함수 활용하기

```python
bins = [0, 100, 200, 300, df_copy['운동량'].max()] # 내가 보고 싶은 값 리스트로 지정, df_copy 운동량의 최댓값도 넣음
pd.cut(df_copy['운동량'], bins) # 내가 입력한 단위 사이에 운동량 있는지 확인 가능
# > 0 (0.0, 100.0] : 0번째 운동량 = 0~100 사이에 있단 것
```

```python
# 자동으로 cut하기
pd.cut(df_copy['운동량'], bins=10) # 최솟값~최댓값 사이 10개 구간으로 나눈 것 >> 근데 극단값 있으면 구간 이상해지므로, 극단값 제거 필요
pd.cut(df_copy['운동량'], bins=10).value_counts()
```

```python
# qcut() : 데이터 균등하지 않을 경우 사용, 균등하면 cut 써도 됨
pd.qcut(df_copy['운동량'], q=10) # 일정 비율대로 나누기
pd.qcut(df_copy['운동량'], q=10).value_counts()
```

```python
# 표준분포로 표기하기: # 이상치 어느 정도 제거하고 싶을 때 이런 식으로 할 수 있음
qcut_bins = [0, 0.2, 0.8, 1] # 0퍼부터 20퍼, 20~80퍼, 80~100퍼 구간으로 나누겠단 것
df_copy['운동량_한글'] = pd.qcut(df_copy['운동량'], qcut_bins, labels=['적음', '보통', '많음'])
    # labels 부여해서, 0퍼~20퍼 사이의 행은 운동량 적은 것 등으로 할당하는 것
df_copy.head()
```
