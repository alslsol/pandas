```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('titanic')
df_copy = df.copy()
```

```python
# 성별 기준으로 생존율 평균 구하기
# 성별 기준으로 그루핑 >> 이후 성별 따라 평균 계산
df_copy.groupby('sex').mean(numeric_only=True) # 숫자만 가져와서 평균 계산
```

```python
# 여러 개의 컬럼 그룹핑: 먼저 성별 기준으로 그룹핑 >> 각 성별 내에서 몇등석인지 기준으로 그룹핑
df_copy.groupby(['sex', 'pclass']).mean(numeric_only=True)
```

```python
# 보고 싶은 정보만 선택하기
df_copy.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only=True)
    # [[]]: 하면 Series를 df 형태로 보여줌
```

```python
# 그루핑해서 선택한 데이터를 >> 인덱스로 풀기
df_temp = df_copy.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only=True)
df_temp
```

```python
df_temp.reset_index()
```

```python
# 여러 연산 동시에 출력
df_copy.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean', 'median'])
```

```python
# 2개 컬럼 간 연관성 확인 위해 pivot_table 활용 >> 교차분석
df_copy.pivot_table(index='who', values = 'survived')
```

```python
df_copy.pivot_table(index=['who', 'pclass'], values='survived')
```
